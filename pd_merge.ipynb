# note 2
There are five kinds of merge? 

- `how = 'inner'` (default)
- `how = 'outer'`
- `how = 'left'`
- `how = 'right'`
- `how= 'cross'`

```python
#inner
pd.merge(
    left=cust_df,
    right=temp_df,
    how='inner')
```

```python
%%time
#left
pd.merge(
    left=cust_df,
    right=temp_df,
    how='left')
```

```python
#right
pd.merge(
    left=cust_df,
    right=temp_df,
    how='right')
```

```python
#outer
pd.merge(
    left=cust_df,
    right=temp_df,
    how='outer')
```

```python
#cross
student_df = pd.DataFrame([['A',60],['B',65],['C',67],['D',76]],
                  columns=['student','Grade'])
student_df
```

```python
tutor_df = pd.DataFrame([[1,90],[2,95],[3,89],[4,99],[5,96]],
                  columns=['tutor','Quality'])
tutor_df
```

```python
combination_df =(
    pd.merge(
    left=student_df,
    right=tutor_df,
    how='cross'
        )
)
combination_df
```

```python
combination_df['diff'] = combination_df.Quality - combination_df.Grade
```

```python
combination_df
```

# Note 3
when using `how='cross'` we don't need to use `on='ColumnName'` becasue there are no columns that we are merging the tables one.


# Note 4
We made sure that the merging columns have the same name on the left and right dataframe, while it is the recommended practice, we can go around it, using `right_on` and `left_on`.

```python
pd.merge(
    left=cust_df,
    right=temp_df,
    how='left',
    on='Date')
```

```python
temp_df.rename(columns={'Date':'date'})
```

```python
pd.merge(
    left=cust_df,
    right=temp_df.rename(columns={'Date':'date'}),
    how='left',
    on='Date')
```

```python
pd.merge(
    left=cust_df,
    right=temp_df.rename(columns={'Date':'date'}),
    how='left',
    left_on='Date',
    right_on='date')
```
